```python
import cv2 as cv
import numpy as np

blank_image = np.zeros((950,1800,3), np.uint8)

BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

RED = (0, 0, 255)
GREEN = (0, 255, 0)
BLUE = (255, 0, 0)

GREY = (190, 190, 190)
YELLOW = (0, 255, 255)
Purple = (238,0,238)

cv.rectangle(blank_image, (0,0), (1800,950), WHITE, cv.FILLED)

def my_function(x,y):
    
  p1 = x+100, y+100
  p2 = x+400, y+400
  p3 = x+200, y+80
  p4 = x+300, y+120
  p5 = x+80, y+200
  p6 = x+120, y+300
  p7 = x+380, y+200
  p8 = x+420, y+300
    
  center0 = x+250, y+75
  axes0 = 25, 25
  angle0 = 0

  center1 = x+180, y+200
  axes1 = 40, 40
  angle1 = 0

  center2 = x+320, y+200
  axes2 = 40, 40
  angle2 = 0

  center3 = x+180, y+200
  axes3 = 1, 1
  angle3 = 0

  center4 = x+320, y+200
  axes4 = 1, 1
  angle4 = 0

  center5 = x+180, y+200
  axes5 = 20, 20
  angle5 = 0

  center6 = x+320, y+200
  axes6 = 20, 20
  angle6 = 0
    
  center100 = x+250, y+75
  axes100 = 38, 38
  angle100 = 0

  center101 = x+250, y+75
  axes101 = 1, 1
  angle101 = 0

  cv.rectangle(blank_image, (x+70,y+400), (x+430,y+700), GREY, cv.FILLED) #身體
  cv.rectangle(blank_image, (x+70,y+400), (x+430,y+700), BLACK, 3)

  cv.rectangle(blank_image, (x+38,y+400), (x+98,y+600), GREY, cv.FILLED) #左手
  cv.rectangle(blank_image, (x+38,y+400), (x+98,y+600), BLACK, 3)
    
  cv.rectangle(blank_image, (x+402,y+400), (x+462,y+600), GREY, cv.FILLED) #右手    
  cv.rectangle(blank_image, (x+402,y+400), (x+462,y+600), BLACK, 3)

  cv.rectangle(blank_image, (x+170,y+700), (x+250,y+900), GREY, cv.FILLED) #左腳
  cv.rectangle(blank_image, (x+170,y+700), (x+250,y+900), BLACK, 3)

  cv.rectangle(blank_image, (x+250,y+700), (x+330,y+900), GREY, cv.FILLED) #右腳
  cv.rectangle(blank_image, (x+250,y+700), (x+330,y+900), BLACK, 3)

  cv.ellipse(blank_image, center0, axes0, angle0, 0, 360, YELLOW, 25)
  cv.ellipse(blank_image, center101, axes101, angle101, 0, 360, YELLOW, 25)
  cv.ellipse(blank_image, center100, axes100, angle100, 0, 360, BLACK, 3)

  cv.rectangle(blank_image, p1, p2, GREY, cv.FILLED)
  cv.rectangle(blank_image, p1, p2, BLACK, 3)

  cv.rectangle(blank_image, p3, p4, Purple, cv.FILLED)
  cv.rectangle(blank_image, p3, p4, BLACK, 3)

  cv.rectangle(blank_image, p5, p6, RED, cv.FILLED)
  cv.rectangle(blank_image, p5, p6, BLACK, 3)

  cv.rectangle(blank_image, p7, p8, RED, cv.FILLED)
  cv.rectangle(blank_image, p7, p8, BLACK, 3)

  cv.ellipse(blank_image, center1, axes1, angle1, 0, 360, BLACK, 3)
  cv.ellipse(blank_image, center2, axes2, angle2, 0, 360, BLACK, 3)

  cv.ellipse(blank_image, center3, axes3, angle3, 0, 360, BLACK, 3)
  cv.ellipse(blank_image, center4, axes4, angle4, 0, 360, BLACK, 3)

  cv.ellipse(blank_image, center5, axes5, angle5, 0, 360, WHITE, 32)
  cv.ellipse(blank_image, center6, axes6, angle6, 0, 360, WHITE, 32)

  pts = [(x+200, y+310), (x+250, y+230), (x+300, y+310)]

  cv.fillPoly(blank_image, np.array([pts]), RED)

  T1, T2, T3 = (x+200, y+310), (x+250, y+230), (x+300, y+310)
  cv.line(blank_image, T1, T2, 50, 4)
  cv.line(blank_image, T2, T3, 50, 4)
  cv.line(blank_image, T3, T1, 50, 4)

  L1, L2, L3, L4, L5, L6, L7 = (x+190, y+350), (x+210, y+370), (x+230, y+350), (x+250, y+370), (x+270, y+350), (x+290, y+370), (x+310, y+350)

  cv.line(blank_image, L1, L2, 50, 4)
  cv.line(blank_image, L2, L3, 50, 4)
  cv.line(blank_image, L3, L4, 50, 4)
  cv.line(blank_image, L4, L5, 50, 4)
  cv.line(blank_image, L5, L6, 50, 4)
  cv.line(blank_image, L6, L7, 50, 4)

my_function(0,0)
my_function(550,0)
my_function(1100,0)

cv.imshow('My Image', blank_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
